Ex4: Write a function called first_diff that is given two strings and returns the first location in
which the strings differ. If the strings are identical, it should return -1.Ignore the case in your function. It means h and H are the same.

```python
def first_diff(str1, str2):

    str1 = str1.lower()
    str2 = str2.lower()
    min_len = min(len(str1), len(str2))

    for i in range(min_len):
        if str1[i] != str2[i]:
            return i

    if len(str1) != len(str2):
        return min_len

    return -1

first_diff("Hello", "hello")
```

Ex6: Write a function that solves the system of equations ax + by = e and cx +dy = f. Your function should takes 6 parameters, (a, b, c, d, e,f) and return the values of x and y.

```python
def solve_system(a, b, c, d, e, f):
    denominator = a*d - b*c
    if denominator == 0:
        return None

    x = (e*d - b*f) / denominator
    y = (a*f - e*c) / denominator
    return x, y
solve_system(4, 2, 1, 2, 8, 6)
```

Ex7:  Write a function called change_case that given a string, returns a string with each upper  case letter replaced by a lower case letter and vice-versa

```python
def change_case(c):
    result = ""
    for char in c:
        if char.isupper():
            result += char.lower()
        elif char.islower():
            result += char.upper()
        else:
            result += char
    return result

change_case("Hello Htoo")
```

Ex8:  Write a function called primes that is given a number n and returns a list of the first n  primes. Let the default value of n be 100.

```python
def primes(n=100):
    prime_list = []
    num = 2  # start from the first prime number

    while len(prime_list) < n:
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                break
        else:

            prime_list.append(num)
        num += 1

    return prime_list

primes(5)
```

Ex9: A Tic-tac-toe board can be represented be a 3×3 two-dimensional list, where zeroes stand for  empty cells, ones stand for X’s and twos stand for O’s. Write a function that is given such a list and randomly chooses a spot in which to place  a 2 (O). The spot chosen must currently be empty and a spot must be chosen.

```python
import random

def place_O(board):

    empty_spots = []
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                empty_spots.append((i, j))


    if empty_spots:
        i, j = random.choice(empty_spots)
        board[i][j] = 2

    return board

board = [
    [0, 0, 2],
    [0, 1, 0],
    [2, 0, 1]
]

new_board = place_O(board)
for row in new_board:
  print(row)
```

Ex10:  Write a function that is given a 9 × 9 potentially solved Sudoku and returns True if it is solved correctly and False if there is a mistake. The Sudoku is correctly solved if there are no repeated numbers in any row or any column or in any of the nine “blocks.”


```python
def is_sudoku_solved(board):

    for row in board:
        seen = []
        for num in row:
            if num in seen:
                return False
            seen.append(num)


    for col in range(9):
        seen = []
        for row in range(9):
            num = board[row][col]
            if num in seen:
                return False
            seen.append(num)


    for block_row in range(0, 9, 3):
        for block_col in range(0, 9, 3):
            seen = []
            for i in range(3):
                for j in range(3):
                    num = board[block_row + i][block_col + j]
                    if num in seen:
                        return False
                    seen.append(num)

    return True

sudoku_board = [
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
]
is_sudoku_solved(sudoku_board)
```
