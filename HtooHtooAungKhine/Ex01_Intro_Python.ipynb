# Ex 4: Daily Maths

You are working in a retail shop. Write a program that computes the total cost for a customer that bought 3 differnt types of fruits : apple, orange, mango. Ask the user to enter the quantity and price for each type.

```python
quant_apple=int(input("Enter the quantity of apple you want: "))
price_apple=float(input("Enter the price of apple you want: "))
total_apple=quant_apple * price_apple

quant_orange=int(input("Enter the quantity of orange you want: "))
price_orange=float(input("Enter the price of orange you want: "))
total_orange=quant_orange * price_orange

quant_mango=int(input("Enter the quantity of mango you want: "))
price_mango=float(input("Enter the price of mango you want: "))
total_mango=quant_mango * price_mango

total_cost= total_apple+total_orange+total_mango

print("Your total amount is", total_cost)
```

# Ex 9: Computing

Write a program that asks the user for a number of seconds and prints out how many minutes and seconds that is. For instance, 200 seconds is 3 minutes and 20 seconds.

```python
time_sec= int(input("Enter the number of seconds: "))
exact_min=time_sec //60
exact_sec=time_sec % 60
print("{} seconds is {} minutes and {} seconds".format(time_sec,exact_min,exact_sec))
```

# Ex 10: Using Math Module

Write a program that asks the user for a number and then prints out the sine, cosine, and tangent of that number.

```python
import math

num = float(input("Enter a number:"))
sine_value = math.sin(num)
cosine_value = math.cos(num)
tangent_value = math.tan(num)
print("Sine value is",sine_value)
print("Cosine value is",cosine_value)
print("Tangent value is",tangent_value)
```
